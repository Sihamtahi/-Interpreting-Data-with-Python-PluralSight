```python
crode_oil_data = pd.read_csv("datasets/U.S._crude_oil_production.csv")

crode_oil_data.shape
crode_oil_data.columns
 
```

```python
crode_oil_data.columns[crode_oil_data.sum(axis=0) == 0]
crode_oil_data.drop(['Arizona','Virginia'],inplace =True, axis =1)
```

```python
crode_oil_data.head()
```

```python
crode_oil_data['Date']= pd.to_datetime(crode_oil_data['Month'])
```

```python
crode_oil_data.head()
```

```python
crode_oil_data.drop(['Month'],inplace =True, axis =1)
crode_oil_data.columns
```

```python
crode_oil_data =crode_oil_data.rename (columns={'Federal Offshore Gulf of Mexico Crude Oil' :'Mexico','Federal Offshore Pacific Crude Oil':'Pacific'})
crode_oil_data.columns
```

```python
crode_oil_data['Year'] =crode_oil_data['Date'].dt.year
crode_oil_data['Year'].sample(10)
```

```python
crode_oil_data['Month'] =crode_oil_data['Date'].dt.month
crode_oil_data['Month'].sample(10)
```

```python
 ####"save this data into new file ###
crode_oil_data.to_csv('datasets/crude_oil_data_processed.csv',index =False)
```
