```python
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt 
from scipy.stats import pearsonr, spearmanr
import numpy as np
```

```python
bikesharing_data = pd.read_csv("datasets/bike_sharing_daily.csv")
bikesharing_data.head()
bikesharing_data.columns
bikesharing_data.shape
```

```python
bikesharing_data.head()
```

```python
np.corrcoef(bikesharing_data['temp'],bikesharing_data['cnt'])
```

```python
 ##Corr based on the linear relachionship between the two variables
pearsonr(bikesharing_data['temp'],bikesharing_data['cnt'])
```

```python
##Corr based on ranks of var rather than the actual raw values
spearmanr(bikesharing_data['temp'],bikesharing_data['cnt'])
##small pvallue for both ---> the value is significant
```

```python
plt.figure(figsize=(12,8))
plt.scatter(bikesharing_data['temp'],bikesharing_data['cnt'],color='m')
plt.title('Bike sharing daily')
plt.xlabel('temperature')
plt.ylabel('count')
plt.show()
```
